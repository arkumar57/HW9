**Description of the problem/decision**

Airbnb serves as a marketplace connecting hosts and guests, offering a variety of listings within neighborhoods for short-term stays. Our goal is to predict the number of reviews a particular listing might receive within a month. This prediction can help gauge the potential popularity of future listings before they go live, enabling hosts to create more appealing and competitive listings.

**Description of the dataset (the raw data and/or some EDA)**

The dataset provided to train the model initially contained 16 features. For better efficiency of the model, we excluded the features that were not relevant for the prediction, such as `Host ID`, `Host Name`, and `ID`. Also, features like `Last Review Date` and the total `Number of Reviews` for a property were removed to avoid data leakage. Data leakage is a situation where the model has access to information to make predictions artificially easy, which would compromise its applicability in the real world. Additionally, there were serious outliers in several of the numerical features ("price", "minimum_nights", and "calculated_host_listings" have a few values which drastically expand the range of the feature). Outliers have a strong effect on linear models, but also make standard scaling less effective.Hence we implemented a logistic scaler on these features. Latitude and longitude features seemed to have some predictive value, but didn't have a linear connection with the target value. Instead, we made them categorical features by binning them.

**Description of the model**

Starting of with the base model for implementation, we applied dummy classifier. Unsurprisingly, the model generated very poor results. Unfortunately, linear models like Ridge also gave poor results. We played around with the alpha value as we expected some issues with outliers but there didn't seem to be a hyperparameter that gives a high test or train score, suggesting that the data would do better with a non-linear model.

Non-linear models were an improvement! Even the basic DecisionTreeRegressor was doing significantly better than Ridge. The test scores were still quite low but clustering showed some moderate success, with both RandomForestRegressor and LGBMRegressor showing test scores of around .4. 

Conclusively, we found LBGMRegressor to be the best performing model onto our data because it has a slightly higher average test score than RandomForestRegressor as well as a lower train score. 

**Description of results**

The results of the study showed some interesting observations. Unsurprisingly, price and minimum nights showed a strong affect on the data, with a high price having a negative affect on reviews per month and high minumum nights also having a negative affect. Surprisingly however, 365 availability has a very large affect! Looking at a few of the data points in a SHAP plot shows some interesting relationships however! It seems that the feature was rated so highly despite not showing a strong predictive value in the scatter plot, was because a score of 0 has a very large negative affect on reviews per month. This is unsurprising but also suggests a general issue with the target value and how it favors certain properties in what is possibly a confounding way.

**Concerns**

While our overall error was not terrible (see figure Above), here are several factors that may have contributed to potentially misleading results. First, the target value, reviews per month, has a substantial issue with how it was calculated. The problem is that this number is generated by dividing total number of reviews by total months open. Because AirB&Bs often have a period of inactivity where they have no guests and thus no reviews, many AirB&Bs that are open for more than a few months have a substantially lower target value than AirB&Bs which have just opened. The graph which compares `Last Review` to `Reviews per Month` below demonstrates this effect, which damages the whole dataset as our target value can be influenced by a feature that really has nothing to do with what makes an AirB&B popular. To minimize the effect of this, we dropped the `Last Review` feature the data we fed into the model, but the effect of this underlying issue could still be a problem with the dataset.
